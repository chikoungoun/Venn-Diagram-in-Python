- call matplolib for graphing functionalities

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

- call the 'venn' library 

```python
import matplotlib_venn as vplt
```

```python
v = vplt.venn2(subsets={'10':2,'01':2,'11':1},set_labels=('A','B'))
plt.show()
```

### Venn 2  : takes a 3 elements list for 2 circle venn diagram

- Subsets

```python
vplt.venn2(subsets=(3,2,1))
```

- Sets

```python
vplt.venn2([set(['A','B','C','D']),set(['D','E','F'])])
```

### Venn 3 : takes a 7 element list of subsets to draw a 3 circle Venn diagram

```python
vplt.venn3(subsets=(1,1,1,2,1,2,2),set_labels=('Set1','Set2','Set3'))
```
